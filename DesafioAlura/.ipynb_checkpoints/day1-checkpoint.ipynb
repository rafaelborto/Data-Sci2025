## RESPOSTA GPT

```python
import requests

# URL da API
url = 'https://last-airbender-api.fly.dev/api/v1/characters'

try:
    # Realiza a requisi√ß√£o GET
    response = requests.get(url)
    response.raise_for_status()  # Verifica se houve erro na requisi√ß√£o

    # Converte a resposta para JSON
    data = response.json()

    # Imprime o corpo da resposta
    print(data)

except requests.exceptions.RequestException as e:
    print(f'Ocorreu um erro na requisi√ß√£o: {e}')
except ValueError:
    print('Erro ao decodificar a resposta JSON.')
```

üîç Explica√ß√£o
<code>requests.get(url):</code> Envia uma requisi√ß√£o HTTP GET para a URL especificada.

<code>response.raise_for_status():</code> Lan√ßa uma exce√ß√£o se a requisi√ß√£o retornou um erro HTTP.

<code>response.json():</code> Converte o conte√∫do da resposta para um objeto Python (geralmente uma lista ou dicion√°rio).

üìå Observa√ß√µes
A API retorna, por padr√£o, 20 personagens por p√°gina. Para obter mais personagens ou navegar entre as p√°ginas, voc√™ pode utilizar os par√¢metros <code>perPage</code> e <code>page</code>. Por exemplo:

url = 'https://last-airbender-api.fly.dev/api/v1/characters?perPage=50&page=2'

Certifique-se de ter o m√≥dulo requests instalado. Caso n√£o tenha, voc√™ pode instal√°-lo usando o comando pip install requests.

## C√ìDIGOS COMBINADOS

```python
import json
import requests

def personagens():
    api_url = 'https://last-airbender-api.fly.dev/api/v1/characters'

    try:
        response = requests.get(api_url)
        response.raise_for_status()  # Verifica se houve erro na requisi√ß√£o

        data = response.json()
        print(json.dumps(data, indent=4))  # Sa√≠da formatada

    except requests.exceptions.RequestException as e:
        print(f'Ocorreu um erro na requisi√ß√£o: {e}')
    except ValueError:
        print('Erro ao decodificar a resposta JSON.')

personagens()
```
