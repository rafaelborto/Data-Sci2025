## MÉTODO .FIND() AVANÇADO

Para contar quantas vezes a sequência 'deo' aparece na variável frase2 = "curso em video pyhton deo", você pode utilizar diferentes abordagens em Python. Aqui estão algumas opções:\
✅ Usando o método <CODE>.count()</CODE>:\
O método count() retorna o número de ocorrências não sobrepostas de uma substring em uma string.


```python
frase2 = "curso em video pyhton deo"
ocorrencias = frase2.count('deo')
print(ocorrencias)  # Saída: 2
```

🔁 Usando um loop com <CODE>.find()</CODE>\
Se você deseja obter os índices de todas as ocorrências da substring, pode usar um loop com o método<CODE>.find()</CODE>:

```python
frase2 = "curso em video pyhton deo"
substring = 'deo'
indices = []
inicio = 0

while True:
    pos = frase2.find(substring, inicio)
    if pos == -1:
        break
    indices.append(pos)
    inicio = pos + len(substring)  # Avança para evitar sobreposição

print(indices)  # Saída: [11, 22]
print(len(indices))  # Saída: 2
```

🧪 Usando expressões regulares com <CODE>re.finditer()</CODE>\
Para uma abordagem mais poderosa, especialmente útil em casos mais complexos, você pode usar o módulo re:\
O <CODE>re.finditer()</CODE> retorna um iterador com todos os matches não sobrepostos, permitindo extrair os índices de início de cada ocorrência.

```python
import re
frase2 = "curso em video pyhton deo"
substring = 'deo'
ocorrencias = [m.start() for m in re.finditer(substring, frase2)]

print(ocorrencias)  # Saída: [11, 22]
print(len(ocorrencias))  # Saída: 2
```
